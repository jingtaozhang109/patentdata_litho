```python
import pandas as pd
```

```python
mask_all_data = pd.read_excel("./data/litho_western_eruope.xlsx")
mask_all_data
```

```python
mask_all_data = mask_all_data.drop_duplicates()
mask_all_data.head()
```

```python
# 提取我们所需要的 GA 和 IP
df_end = mask_all_data[[' GA ',' IP ']]
# 对年进行提取 先将数值转换为字符串 然后去掉前后的空格再提取前四位字符
df_end['year'] = df_end[' GA '].apply(lambda x:str(x).strip()[:4])
# 将 IP 列转换为列表 然后将其转换为多行数据 每行数据只有一个 IPC 
df_end['IP'] = df_end[' IP '].apply(lambda x:str(x).split(";"))
# 提取处理之后的 IPC 和 年份
df_end = df_end[['IP','year']]
# 删除空值
df_end  = df_end.dropna()
# 将列表转换为多行值
df_end = df_end.explode('IP')
# 去掉 IP 里面的空格
df_end['IP'] = df_end['IP'].apply(lambda x: x.strip())
# 重置 index
df_end  = df_end.reset_index(drop=True)
df_end['IP3'] = df_end['IP'].apply(lambda x:x.split('-')[0].strip()[:3])
df_end['IP4'] = df_end['IP'].apply(lambda x:x.split('-')[0].strip())
df_end
```

```python
df_end.to_csv("./data/litho_western_eruope_deal.csv",index=False)
```
