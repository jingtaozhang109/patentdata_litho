```python
# 关于某个年份 双方专利子类别的重叠 前四位
overlap_4 = 0
for one in getyeardram['IP4']:
    for two in getyearlitho['IP4']:
        if(one == two):
            overlap_4 += 1 
print(overlap_4)
```

```python
# 关于某个年份 双方专利数量总和
sumOfTheTwo = len(getyeardram) + len(getyearlitho) 
sumOfTheTwo
```

```python
# 关于某个年份 双方专利子类别的重叠 前三位
overlap_3 = 0
for one in getyeardram['IP3']:
    for two in getyearlitho['IP3']:
        if(one == two):
            overlap_3 += 1
print(overlap_3)
```

```python
# 关于某个年份 光刻企业前四位专利重叠数量
overlap_litho = 0
litho_ip4_ls = getyearlitho['IP4'].tolist()
# 第一个从0开始 第二个就从1 开始
# 第一个从1开始 第二个就从2 开始
for one in range(len(litho_ip4_ls)):
    for two  in range(one+1 ,len(litho_ip4_ls)):
        if litho_ip4_ls[one] == litho_ip4_ls[two]:
            overlap_litho +=1
print(overlap_litho)
```

```python
# 关于某个年份 光刻企业专利数量总和
sumOflitho = len(getyearlitho) 
sumOflitho
```

```python
# 封装成函数 可以获取不同年份的q1的值
def get_year_data(year):
    # 如果有一个的数据获取不到 就返回0
    try:
        getyeardram = data1_dram.get_group(year) # 获取21年DRAM的数据
        getyearlitho= data1_litho.get_group(year) # 获取21年光刻技术的数据
        # 关于某个年份 双方专利子类别的重叠 前四位
        overlap_4 = 0
        for one in getyeardram['IP4']:
            for two in getyearlitho['IP4']:
                if(one == two):
                    overlap_4 += 1
        # 关于某个年份 双方专利数量总和
        sumOfTheTwo = len(getyeardram) + len(getyearlitho)
        # 关于某个年份 双方专利子类别的重叠 前三位
        overlap_3 = 0
        for one in getyeardram['IP3']:
            for two in getyearlitho['IP3']:
                if(one == two):
                    overlap_3 += 1
        # 关于某个年份 光刻企业前四位专利重叠数量
        overlap_litho = 0
        litho_ip4_ls = getyearlitho['IP4'].tolist()
        # 第一个从0开始 第二个就从1 开始
        # 第一个从1开始 第二个就从2 开始
        for one in range(len(litho_ip4_ls)):
            for two  in range(one+1 ,len(litho_ip4_ls)):
                if litho_ip4_ls[one] == litho_ip4_ls[two]:
                    overlap_litho +=1
        # 关于某个年份 光刻企业专利数量总和
        sumOflitho = len(getyearlitho) 

        data = (overlap_4/sumOfTheTwo) - ((overlap_3/sumOfTheTwo) * (overlap_litho / sumOflitho))
    except:
        data = 0
    return data
data_q1_end = [] # 存储q1的所有年份的数据
for one_year in q1_year_ls:
    data_q1_end.append([one_year,get_year_data(one_year)])
data_q1_end = pd.DataFrame(data_q1_end,columns=['year','技术互补性'])
data_q1_end.index = data_q1_end.year
data_q1_end.head()
```

```python
# 封装成函数 可以获取不同年份的q1的值
def get_year_dataq2(year):
    # 如果有一个的数据获取不到 就返回0
    try:
        getyearmask = data1_mask.get_group(year) # 获取21年mask的数据
        getyearlitho= data1_litho.get_group(year) # 获取21年光刻技术的数据
        # 关于某个年份 双方专利子类别的重叠 前三位代码
        overlap_3 = 0
        for one in getyearmask['IP3']:
            for two in getyearlitho['IP3']:
                if(one == two):
                    overlap_3 += 1
        # 关于某个年份 双方专利数量总和
        sumOfTheTwo = len(getyearmask) + len(getyearlitho)

        # 关于某个年份 光刻企业前四位专利重叠数量
        overlap_litho = 0
        litho_ip4_ls = getyearlitho['IP4'].tolist()
        # 第一个从0开始 第二个就从1 开始
        # 第一个从1开始 第二个就从2 开始
        for one in range(len(litho_ip4_ls)):
            for two  in range(one+1 ,len(litho_ip4_ls)):
                if litho_ip4_ls[one] == litho_ip4_ls[two]:
                    overlap_litho +=1
        # 关于某个年份 光刻企业专利数量总和
        sumOflitho = len(getyearlitho) 

        data = (overlap_3/sumOfTheTwo)   * (overlap_litho / sumOflitho)
    except:
        data = 0
    return data
data_q2_end = [] # 存储q1的所有年份的数据
for one_year in q1_year_ls:
    data_q2_end.append([one_year,get_year_dataq2(one_year)])
data_q2_end = pd.DataFrame(data_q2_end,columns=['year','技术相似性'])
data_q2_end
```

```python
def get_t3(df):
    # 检查 index 是否连续
    is_continuous = pd.RangeIndex(int(df.index.min()), int(df.index.max()) + 1).equals(df.index)

    # 添加缺失的年份
    if not is_continuous:
        start_year = int(df.index.min())
        end_year = int(df.index.max())
        missing_years = [y for y in range(start_year, end_year + 1) if  y not in df.index.tolist()]
        missing_data = pd.DataFrame({'data': [0] * len(missing_years)}, index=missing_years)
        df = pd.concat([df, missing_data]).sort_index()
        print(missing_years)
    t3_data = []
    for one in range(len(df)):
        if(one==0):
            sum_all = 0
        elif(one==1):
            sum_all = df.iloc[0]['data']
        else:
            sum_all = df.iloc[one-1]['data'] + df.iloc[one-2]['data']
        t3_data.append(sum_all)
    df['t3'] = t3_data
    df = df.drop(['data'],axis=1)
    return df
data_mask_t3 = get_t3(data_mask_group_year)
data_mask_t3.head()
```

```python
def get_t5(df):
    # 检查 index 是否连续
    is_continuous = pd.RangeIndex(int(df.index.min()), int(df.index.max()) + 1).equals(df.index)

    # 添加缺失的年份
    if not is_continuous:
        start_year = int(df.index.min())
        end_year = int(df.index.max())
        print(start_year,end_year)
        print(df.index.tolist())
        missing_years = [y for y in range(start_year, end_year + 1) if int(y) not in df.index.tolist()]
        missing_data = pd.DataFrame({'data': [0] * len(missing_years)}, index=missing_years)
        print(missing_years)
        df = pd.concat([df, missing_data]).sort_index()

    t5_data = []
    for one in range(len(df)):
        if(one==0):
            sum_all = 0
        elif(one==1):
            sum_all = df.iloc[0]['data']
        elif(one==2):
            sum_all = df.iloc[0]['data'] + df.iloc[1]['data']  
        elif(one==3):
            sum_all = df.iloc[0]['data'] + df.iloc[1]['data'] + df.iloc[2]['data']
        elif(one==4):
            sum_all = df.iloc[0]['data'] + df.iloc[1]['data'] + df.iloc[2]['data'] + df.iloc[3]['data']
        else:
            sum_all = df.iloc[one-1]['data'] + df.iloc[one-2]['data'] +df.iloc[one-3]['data'] + df.iloc[one-4]['data'] + df.iloc[one-5]['data']
        t5_data.append(sum_all)
    df['t5'] = t5_data
    df = df.drop(['data'],axis=1)
    return df
```
